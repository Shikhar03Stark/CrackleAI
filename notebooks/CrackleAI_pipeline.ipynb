# ASR Model from Hugging face

```python
from transformers import AutoTokenizer, AutoModelForCTC, pipeline, Wav2Vec2FeatureExtractor
tokenizer = AutoTokenizer.from_pretrained("facebook/wav2vec2-large-robust-ft-swbd-300h")
feature_extractor = Wav2Vec2FeatureExtractor.from_pretrained("facebook/wav2vec2-large-robust-ft-swbd-300h")  
model = AutoModelForCTC.from_pretrained("facebook/wav2vec2-large-robust-ft-swbd-300h")
wav2vec2 = pipeline(task="automatic-speech-recognition", model=model, feature_extractor=feature_extractor, tokenizer=tokenizer)
```

# Sentence embedding vectors based on transformers

```python
sentence_embedding = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
```

 # Title Generation t5 based on BERT

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("Michau/t5-base-en-generate-headline")
model = AutoModelForSeq2SeqLM.from_pretrained("Michau/t5-base-en-generate-headline")
title_generator = pipeline("text2text-generation", model=model, tokenizer=tokenizer) 
```

```python
from pydub import AudioSegment
from pydub.silence import split_on_silence

def convertToWav(filename):
    sound = AudioSegment.from_file(filename)
    basename, ext = ''.join(filename.split('.')[:-1]), filename.split('.')[-1]
    # basename = basename.replace('\\\\', '\\')
    sound.export(basename + '.wav', format='wav')
    return basename + '.wav'

def create_audio_chunks(filename, duration_s=10):
  sound = AudioSegment.from_file(filename, format="wav")
  audio_chunks = split_on_silence(sound, min_silence_len=600, silence_thresh=-33)
  result = []
  chunk = 0
  duration = duration_s*1000 #milliseconds
  for audio in audio_chunks:
    start = 0
    if chunk > 1000:
      break
    while start < len(audio):
      end = start + duration
      if end > len(audio):
        end = len(audio)
      chunk += 1
      exp_filename = filename.replace(".wav", "_chunk{0}.wav".format(chunk))
      chunk_sound = audio[start:end]
      # print("exporting", exp_filename)
      chunk_sound.export(exp_filename, format="wav")
      result.append(exp_filename)
      start = end
  return result

def forward_audio(chunk_path, model=wav2vec2):
  return model(chunk_path)
```

 ## Topic Segmentation Algorithm

```python
def cosine_similarity(u, v):
  return dot(u, v)/(norm(u)*norm(v))


def weighted_avg(n, a, b):
  return (n*a + b)/(n+1)

def topic_segmentation(sentences, model):
  partitions = []
  part_emb = np.zeros((384,))
  partition = []
  for sent in sentences:
    if len(partition) == 0:
      part_emb = model.encode(sent)
      partition.append(sent)
    else:
      e = model.encode(sent)
      cosine = cosine_similarity(part_emb, e)
      if cosine > 0.25:
        #Extend partion
        partition.append(sent)
        part_emb = weighted_avg(len(partition), part_emb, e)
      else:
        #New Partition
        partitions.append(partition)
        partition = [sent]
        part_emb = e

  if len(partition) > 0:
    partitions.append(partition)

  return partitions
  
```

## Title generation

```python
def generate_title(content, model):
  text = ' '.join(content)
  if( len(text) < 5):
    return text
  title = model(text)[-1]['generated_text']
  # print("fn", model(text))
  return title
```

## Usage

 - Convert video to .wav format
 - Create small chunks of audio for easier processing

```python
audio_path = convertToWav("D:/dev/Cracke-AI/Data/videoplayback_demo.mp4")
audio_chunks = create_audio_chunks(audio_path)
```

```python
print("Number of audio chunks", len(audio_chunks))
```

 - Perform ASR on each chunk

```python
transcript = []
for chunk in audio_chunks:
  transcript.append(forward_audio(chunk, wav2vec2)['text'])
  print(transcript[-1])
```

```python
topics = topic_segmentation(transcript, sentence_embedding)
print("Number of topics extracted", len(topics))
```

```python
output = []
for topic in topics:
  title = generate_title(topic, title_generator)
  output.append((title, topic))
  print(title)
```
